```python
import graphlab
```

#Load the CIFAR-10 dataset

```python
image_train = graphlab.SFrame('image_train_data/')
```

```python
image_train.head()
```

```python
knn_model = graphlab.nearest_neighbors.create(image_train,features=['deep_features'], 
                                              label='id')
```

```python
graphlab.canvas.set_target('ipynb')
cat = image_train[18:19]
cat['image'].show()
```

```python
knn_model.query(cat)
```

```python
def get_images_from_ids(query_result):
    return image_train.filter_by(query_result['reference_label'], 'id')
```

```python
cat_neigbors = get_images_from_ids( knn_model.query(cat) )
```

```python
cat_neigbors['image'].show()
```

```python
car = image_train[8:9]
car['image'].show()
```

```python
get_images_from_ids(knn_model.query(car))['image'].show()
```

# Justa for fun, lets create lambda to find nd show nearest neighbor images

```python
show_neihbors = lambda i : get_images_from_ids(knn_model.query(image_train[i:i+1]))['image'].show()
```

```python
show_neihbors(2)
```

```python
show_neihbors(1)
```
